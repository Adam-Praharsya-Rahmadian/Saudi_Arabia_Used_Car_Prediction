### **Check Multicolinearity**

Multikolinearitas adalah kondisi di mana dua atau lebih variabel independen dalam model regresi linear berganda memiliki korelasi yang sangat tinggi. Korelasi yang berlebihan antar variabel bebas ini dapat menimbulkan masalah dalam model regresi yang akan dibangun, terutama dalam menginterpretasikan kontribusi masing-masing variabel. Oleh karena itu, penting untuk mengevaluasi apakah terdapat korelasi tinggi antara fitur-fitur dalam dataset yang dapat memengaruhi akurasi dan stabilitas model.

Untuk mengidentifikasi adanya multikolinearitas, kita dapat menggunakan metode Variance Inflation Factor (VIF). VIF mengukur seberapa besar variabilitas suatu koefisien regresi dipengaruhi oleh variabel independen lainnya. Jika nilai VIF suatu variabel terlalu tinggi, maka kemungkinan besar variabel tersebut memiliki multikolinearitas yang signifikan dengan variabel lain.

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

def calculate_vif(df):
    vif_data = pd.DataFrame()
    vif_data["Variable"] = df.columns
    vif_data["VIF"] = [variance_inflation_factor(df.values, i) for i in range(df.shape[1])]
    return vif_data
```

```python
calculate_vif(df[['Year','Engine_Size','Mileage']])
```

Hasil perhitungan **Variance Inflation Factor (VIF)** menunjukkan:

- **`Year`** : **6.30**
- **`Engine_Size`** : **5.88**
- **`Mileage`** : **2.52**

Karena semua nilai **VIF < 10**, tidak ada indikasi multikolinearitas yang signifikan. Oleh karena itu, fitur-fitur ini tetap dapat digunakan dalam model tanpa perlu dilakukan penghapusan atau modifikasi lebih lanjut.

#### **Evaluasi Model dengan Metrik**  

- **RMSE (Root Mean Squared Error)**  
  RMSE mengukur selisih antara nilai prediksi dan nilai observasi yang sebenarnya. Semakin kecil nilai RMSE, semakin akurat model dalam melakukan prediksi. Model yang memiliki nilai RMSE rendah diharapkan mampu menghasilkan estimasi harga mobil dengan tingkat kesalahan yang minim.  

- **MAE (Mean Absolute Error)**  
  MAE menghitung rata-rata selisih absolut antara nilai prediksi dan nilai sebenarnya. Nilai MAE yang lebih kecil menunjukkan bahwa model memiliki tingkat kesalahan yang lebih rendah dalam memprediksi harga mobil. Dengan kata lain, model dengan MAE yang rendah dianggap lebih akurat dalam estimasi harga.  

- **MAPE (Mean Absolute Percentage Error)**  
  MAPE menyatakan kesalahan absolut dalam bentuk persentase, sehingga memungkinkan kita untuk memahami skala kesalahan prediksi dibandingkan dengan rentang data. Semakin kecil nilai MAPE, semakin baik performa model dalam menghasilkan prediksi dengan tingkat kesalahan yang rendah.

```python
# Funsi Model Cross Validation

score_rmse = []
nilai_mean_rmse = []
nilai_std_rmse = []

score_mae = []
nilai_mean_mae = []
nilai_std_mae = []

score_mape = []
nilai_mean_mape = []
nilai_std_mape = []

crossval = KFold(n_splits=5, shuffle=True, random_state=42)

for model in models:
    estimator = Pipeline([
        ('Preprocessing', transformer),
        ('model', model)
    ])

    # RMSE
    model_cv_rmse = cross_val_score(
        estimator,
        X_train,
        y_train,
        cv=crossval,
        scoring='neg_root_mean_squared_error',
        error_score='raise'
    )
    score_rmse.append(model_cv_rmse)
    nilai_mean_rmse.append(-model_cv_rmse.mean())  # Negate the values for RMSE
    nilai_std_rmse.append(model_cv_rmse.std())

    # MAE
    model_cv_mae = cross_val_score(
        estimator,
        X_train,
        y_train,
        cv=crossval,
        scoring='neg_mean_absolute_error',
        error_score='raise'
    )
    score_mae.append(model_cv_mae)
    nilai_mean_mae.append(-model_cv_mae.mean())  # Negate the values for MAE
    nilai_std_mae.append(model_cv_mae.std())

    # MAPE
    model_cv_mape = cross_val_score(
        estimator,
        X_train,
        y_train,
        cv=crossval,
        scoring='neg_mean_absolute_percentage_error',
        error_score='raise'
    )
    score_mape.append(model_cv_mape)
    nilai_mean_mape.append(-model_cv_mape.mean())  # Negate the values for MAPE
    nilai_std_mape.append(model_cv_mape.std())
```

# **Cara Kerja XGBoost**

XGBoost (**Extreme Gradient Boosting**) adalah algoritma machine learning berbasis **ensemble learning** yang menggabungkan beberapa pohon keputusan (**decision trees**) untuk meningkatkan akurasi prediksi. Dibandingkan dengan model boosting lainnya, XGBoost lebih cepat dan efisien karena mengoptimalkan proses pembelajaran dengan teknik paralelisasi dan regularisasi.

---

## **Tahapan Cara Kerja XGBoost**

### **1. Inisialisasi Model**
- XGBoost memulai dengan model dasar yang biasanya berupa **prediksi awal sederhana**, seperti rata-rata dari target dalam kasus regresi.

### **2. Membangun Pohon Keputusan Secara Bertahap**
- Model membuat **pohon keputusan pertama** untuk memprediksi nilai target.
- Selanjutnya, pohon berikutnya dibuat untuk **memperbaiki kesalahan** dari pohon sebelumnya.
- Proses ini dilakukan secara **iteratif**, di mana setiap pohon baru belajar dari error yang masih ada.

### **3. Menggunakan Residual Error**
- Residual error adalah **selisih antara prediksi dan nilai sebenarnya**.
- Setiap pohon baru dilatih untuk **memprediksi residual error**, bukan langsung memprediksi nilai target.
- Dengan demikian, model terus menyempurnakan prediksi hingga error semakin kecil.

### **4. Pembobotan dan Penggabungan Pohon**
- Setiap pohon memiliki **bobot kontribusi** yang diatur menggunakan parameter *learning rate*.
- Prediksi akhir diperoleh dengan **menjumlahkan kontribusi dari semua pohon** yang telah dilatih.

### **5. Regularisasi untuk Mencegah Overfitting**
- XGBoost menggunakan **L1 dan L2 regularization** untuk menghindari overfitting dengan mengontrol kompleksitas model.
- **Pruning (pemangkasan pohon)** diterapkan agar model tidak terlalu kompleks.

---

## **3. Keunggulan XGBoost**
- **Cepat dan Efisien** → Menggunakan teknik paralelisasi untuk mempercepat proses training.  
- **Menangani Missing Values** → Secara otomatis menangani data yang hilang tanpa perlu imputasi manual.  
- **Menggunakan Regularisasi** → Mengurangi risiko overfitting dengan teknik L1 dan L2.  
- **Dapat Menangani Data Kategorikal** → Bisa digunakan dengan One-Hot Encoding atau metode khusus.  
- **Mampu Menyesuaikan Kompleksitas Model** → Dengan parameter seperti *max_depth* dan *learning rate*.  

---

## **4. Kesimpulan**
XGBoost adalah salah satu algoritma terbaik untuk **prediksi harga mobil bekas**, karena mampu:
- **Meningkatkan akurasi** dengan pendekatan boosting yang mengoptimalkan setiap iterasi.
- **Mengurangi error** melalui pemanfaatan residual dalam setiap langkah pembelajaran.
- **Bekerja dengan efisien** bahkan pada dataset besar dengan banyak fitur.

Dengan keunggulannya, XGBoost sering digunakan dalam kompetisi **machine learning** dan menjadi pilihan utama untuk model prediksi yang kompleks.

### **Hyperparameter Tuning XGBoost**
Proses tuning pada model XGBoost bertujuan untuk meningkatkan kinerja model yang telah ditentukan sebelumnya. Dengan melakukan optimasi hyperparameter, diharapkan model dapat bekerja lebih efisien dan menghasilkan prediksi yang lebih akurat.

Lalu, apa saja yang perlu diperhatikan dalam melakukan tuning hyperparameter pada XGBoost? Beberapa aspek penting yang harus diperhatikan meliputi pemilihan parameter yang tepat, seperti learning rate, max depth, n_estimators, gamma, serta teknik tuning yang digunakan, seperti Grid Search, Random Search, atau Bayesian Optimization untuk menemukan kombinasi terbaik.

### **1. Learning Rate**
Menentukan besarnya perubahan yang diterapkan pada model di setiap iterasi.  
- **Nilai terlalu tinggi** → Risiko overfitting meningkat.  
- **Nilai terlalu rendah** → Model bisa mengalami underfitting.  
- **Rentang umum**: 0.01 - 0.1.  
- **Efek**: Learning rate yang lebih kecil membuat model lebih stabil tetapi membutuhkan waktu lebih lama untuk mencapai akurasi optimal.  

### **2. n_estimators**  
Menentukan jumlah pohon dalam model boosting.  
- **Nilai terlalu kecil** → Model kurang akurat.  
- **Nilai terlalu besar** → Risiko overfitting meningkat.  
- **Rentang umum**: 100 - 1000.  
- **Efek**: Semakin tinggi nilainya, semakin baik akurasi tetapi dengan risiko overfitting yang lebih besar.  

### **3. max_depth**  
Menentukan kedalaman maksimum pohon dalam model.  
- **Nilai terlalu tinggi** → Overfitting.  
- **Nilai terlalu rendah** → Model kurang akurat.  
- **Rentang umum**: 3 - 10.  
- **Efek**: Kedalaman pohon yang lebih besar meningkatkan kompleksitas dan akurasi tetapi juga meningkatkan risiko overfitting.  

### **4. min_child_weight**  
Menentukan jumlah minimum data yang harus ada di setiap daun pohon.  
- **Nilai terlalu tinggi** → Underfitting.  
- **Nilai terlalu rendah** → Overfitting.  
- **Rentang umum**: 1 - 5.  
- **Efek**: Nilai yang lebih tinggi membuat model lebih stabil tetapi dapat mengurangi akurasi prediksi.  

### **5. gamma**  
Menentukan ambang batas untuk pemisahan node dalam pohon.  
- **Nilai terlalu tinggi** → Underfitting.  
- **Nilai terlalu rendah** → Overfitting.  
- **Rentang umum**: 0.01 - 0.2.  
- **Efek**: Semakin tinggi nilai gamma, semakin selektif model dalam membagi node, yang bisa meningkatkan generalisasi tetapi juga mengurangi kompleksitas model.  

### **6. subsample**  
Menentukan persentase data yang digunakan untuk membangun setiap pohon.  
- **Nilai terlalu rendah** → Underfitting.  
- **Nilai terlalu tinggi** → Overfitting.  
- **Rentang umum**: 0.8 - 1.0.  
- **Efek**: Subsampling membantu mengurangi overfitting dengan memastikan setiap pohon tidak melihat seluruh data dalam satu iterasi.  

### **7. colsample_bytree**  
Menentukan proporsi fitur yang digunakan untuk membangun setiap pohon.  
- **Nilai terlalu rendah** → Underfitting.  
- **Nilai terlalu tinggi** → Overfitting.  
- **Rentang umum**: 0.8 - 1.0.  
- **Efek**: Mengontrol jumlah fitur yang digunakan di setiap pohon untuk meningkatkan efisiensi dan mengurangi risiko overfitting.
